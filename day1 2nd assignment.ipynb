```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpt
```

```python
days = [1,2,3,4,5,6,7]
sales_one = [160, 150, 140, 145, 175, 165, 180]
sales_two = [70, 90, 160, 150, 140, 145, 175]

plt.plot(days, sales_one,'bo--',label='sales one')
plt.plot(days, sales_two,'ro--',label='sales two',alpha=0.70)
plt.xlabel('days')
plt.ylabel('sales')
plt.title('sales graph')
plt.legend(shadow=True)
plt.grid()
plt.show()

```
