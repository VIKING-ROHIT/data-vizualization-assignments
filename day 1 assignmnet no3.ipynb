```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplots(figsize=(8,8))
plt.subplot(2,3,1)
plt.plot(x,y1,label='y1',color='blue',marker='.',linewidth=0.5,alpha=1)
plt.title('subplot 1')
plt.legend(shadow=True)
plt.subplot(2,3,2)
plt.plot(x,y2,label='y2',color='red',marker='.',linewidth=0.5,alpha=1)
plt.title('subplot 2')
plt.legend(shadow=True)
plt.subplot(2,3,3)
plt.plot(x,y3,label='y3',color='yellow',marker='.',linewidth=0.5,alpha=1)
plt.title('subplot 3')
plt.legend(shadow=True)
plt.subplot(2,3,4)
plt.plot(x,y4,label='y4',color='orange',marker='.',linewidth=0.5,alpha=1)
plt.title("subplot4")
plt.legend(shadow=True)
plt.subplot(2,3,5)
plt.plot(x,y5,label='y5',color='black',marker='.',linewidth=0.5,alpha=1)
plt.title('subplot 5')
plt.legend(shadow=True)
```
